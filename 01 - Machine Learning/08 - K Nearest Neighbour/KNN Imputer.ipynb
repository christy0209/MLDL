```python
datanew=pd.get_dummies(datanew)
```

```python
colnames=datanew.columns
colnames
```

```python
from sklearn.impute import KNNImputer #importing knn imputer function
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
data_filled = imputer.fit_transform(datanew)
```
